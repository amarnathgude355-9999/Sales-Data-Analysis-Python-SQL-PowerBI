```python
Sales['order_date'] = pd.to_datetime(Sales['order_date'])
```

Coverting the Column names in  title format

```python
Sales.columns = Sales.columns.str.title()
```

```python
Sales.head()
```

Creating a New column Revenue

```python
Sales['Revenue'] = Sales['Quantity'] * Sales['Price']
```

```python
Sales.head()
```

```python
Sales['Revenue'].dtype
```

Understand the Size of Business

```python
#Cotunt number of rows that means wec an Know the number of orders (business Size)
Sales.shape[0]
```

Total business value?

```python
print(Sales.Revenue.sum())
```

Average Order Value (AOV)

```python
print(Sales['Revenue'].mean())
```

Product-wise Revenue

```python
Sales.groupby('Product')['Revenue'].sum().sort_values(ascending = False)
```

Region-wise Revenue

```python
Sales.groupby('Region')['Revenue'].sum().sort_values(ascending = False)
```

Monthly Revenue Trend (Very Important)

```python
Sales['Month'] = Sales['Order_Date'].dt.to_period('M')
```

```python
Sales
```

```python
Sales.groupby('Month')['Revenue'].sum()
```
